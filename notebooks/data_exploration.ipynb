```python

import torch

import matplotlib.pyplot as plt

import torchvision.transforms as transforms
from src.backend.util.dataloader import get_data_loaders

device = "cuda" if torch.cuda.is_available() else "cpu"
print("Using device: {}".format(device))
```

```python
# Load data
train_dir = '../data/train'
valid_dir = '../data/valid'
test_dir = '../data/test'
batch_size = 32
transform = transforms.Compose([
    transforms.Resize((128, 128)),
    transforms.ToTensor(),
])

train_loader, valid_loader, test_loader, target_to_class = get_data_loaders(
    train_dir, valid_dir, test_dir, batch_size, transform
)
```

```python
def imshow(img):
    # Unnormalize the image if normalization was applied during transformation
    # img = img * std + mean (if you normalized your dataset, adjust accordingly)
    img = img.numpy().transpose((1, 2, 0))  # Convert from Tensor image
    plt.imshow(img)
    plt.show()

# Get a single batch of images
images, labels = next(iter(train_loader))
# Choose a random index
index = torch.randint(len(images), (1,)).item()

# Display the image
imshow(images[index])
class_name = target_to_class[labels[index].item()]
# Optionally, print the label
print('Label:', labels[index])
```
